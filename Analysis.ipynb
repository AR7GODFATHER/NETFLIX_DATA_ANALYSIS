```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
dp=pd.read_csv("final dataset.csv")
md=dp[dp['type']=="Movie"]
td=dp[dp['type']=="TV Show"]
size=[len(md),len(td)]
typ=["Movies","TV Shows"]
plt.bar(typ,size,label="No of movies vs Tv shows",width=0.5)
plt.xlabel("Type")
plt.ylabel("No of Movies and TV Shows")
plt.title("Comparison of no of movies vs no of Tv shows")
plt.legend()
plt.tight_layout()
plt.savefig('Movie vs Tv Shows.png',dpi=300,bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
dp=pd.read_csv("final dataset.csv")
countries=dp['country'].value_counts().head(10)
plt.barh(countries.index,countries.values,label="Countrywise Release Comprison",color="orange")
plt.title("Top 10 countries to release movies")
plt.xlabel("No of Movies")
plt.ylabel("Countries")
plt.legend()
plt.tight_layout()
plt.savefig("Top 10 Movie Countries.png",dpi=300,bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
dp=pd.read_csv("final dataset.csv")
Dur=dp[dp['type'] == "Movie"]['duration'].str.replace(" min",'').astype(int)
plt.hist(Dur,bins=20,color="green",edgecolor='black')
plt.title("Duration wise Movie Comparison")
plt.xlabel("Minutes")
plt.ylabel("No of Movies")
plt.tight_layout()
plt.savefig("Duration comparison.png",dpi=300,bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
dp=pd.read_csv("final dataset.csv")
md=dp[dp['type']=="Movie"]
td=dp[dp['type']=="TV Show"]
trating=td['rating'].value_counts()
mrating=md['rating'].value_counts()
fig, ax=plt.subplots(1,2,figsize=(10,5))
ax[0].pie(mrating.values,labels=mrating.index,autopct='%1.1f%%',pctdistance=0.5)
ax[0].set_title("movie ratings")
ax[1].pie(trating.values,labels=trating.index,autopct='%1.1f%%',pctdistance=0.5)
ax[1].set_title("TV Show ratings")
fig.suptitle("Rating comparison for movies and Tv shows")
plt.tight_layout()
plt.savefig("Ratingpie.png",dpi=300,bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
dp=pd.read_csv("final dataset.csv")
md=dp[dp['type']=="Movie"]
td=dp[dp['type']=="TV Show"]
mgenre=md['listed_in'].value_counts().head(10)
tgenre=td['listed_in'].value_counts().head(10)
fig, ax=plt.subplots(2,1,figsize=(20,10))
ax[0].barh(mgenre.index,mgenre.values)
ax[0].set_xlabel("no of movies")
ax[0].set_ylabel("genres")
ax[0].set_title("genre vs no of movies")
ax[1].barh(tgenre.index,tgenre.values,color="purple")
ax[1].set_xlabel("no of TV Shows")
ax[1].set_ylabel("genres")
ax[1].set_title("genre vs no of TV Shows")
fig.suptitle("Genre vs content")
fig.subplots_adjust(hspace=0.3)
plt.savefig("genrecomp.png",dpi=300,bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
dp=pd.read_csv("final dataset.csv")
md=dp[dp['type']=="Movie"]
td=dp[dp['type']=="TV Show"]
mdat=md['release_year'].value_counts()
tdat=td['release_year'].value_counts()
fig, ax=plt.subplots(2,1,figsize=(15,5))
ax[0].scatter(mdat.index,mdat.values,color='red',marker='o')
ax[0].set_title("No of movies vs release year")
ax[0].grid()
ax[1].scatter(tdat.index,tdat.values,color='green')
ax[1].set_title("No of Tv Shows vs release year")
fig.suptitle("No of content vs Release year",fontsize=14,color="purple",fontweight="bold")
fig.subplots_adjust(hspace=0.5)
ax[1].grid()
plt.savefig("yearwise_release.png",dpi=300,bbox_inches='tight')
plt.show()
```
